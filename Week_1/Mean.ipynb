Python lists are very flexible, but they are slow for big calculations.

**NumPy** arrays can store purely numerical data in much less space, and are much simpler and faster for calculations.

We can calculate the mean with a NumPy array instead of a list:

```
import numpy as np   # numpy is a package which has a set of functions 
# Here we are using variable np to call functions in numpy package
# You can check numpy documentation to see more functions

fluxes = np.array([23.3, 42.1, 2.0, -3.2, 55.6])
m = np.mean(fluxes)
print(m)
```

NumPy has a great range of numerical functions. For example, to calculate the size of an array, and the standard deviation:

```
import numpy as np
fluxes = np.array([23.3, 42.1, 2.0, -3.2, 55.6])
print(np.size(fluxes)) # length of array
print(np.std(fluxes))  # standard deviation
```

Tables(Data) are often stored in comma-separated values (CSV) format. You can use Python's built-in string functions to read a CSV file into a list and process it.

Our file has several rows and columns. We want to store each row in a list and the whole file as a list of these rows.

The program loops over each line in the file, splitting the row into a list of values, and appending each row to data:

```
data = []
for line in open('Week_1_data_1.csv'):      # Loads each line into list variable data 
  data.append(line.strip().split(','))  # Looks for ',' to separate

print(data)
```

Now we can store the data in lists,You would have noticed that they are stored as strings.We need to convert each item from a string to a float. We can't directly convert the entire nested list. We could do this using nested for loops:

```
data = []
for line in open('Week_1_data_1.csv'):        #Line stores each row from the data file
  row = []
  for col in line.strip().split(','):
    row.append(float(col))           #Takes each element and convert it into a float
  data.append(row)

print(data)
```

NumPy has a **simpler** asarray function to do this conversion:

```
import numpy as np

data = []
for line in open('Week_1_data_1.csv'):
  data.append(line.strip().split(','))

data = np.asarray(data, float)
print(data)
```
