#**Introduction to Regression**

Regression takes a group of random variables, thought to be predicting Y, and tries to find a mathematical relationship between them. This relationship is typically in the form of a straight line (linear regression) that best approximates all the individual data points. You have done a similar task in week three with polynimial fit.

In the example below. We first create a dataset with columns Product, cost and profit. Our objective is to predict profit based on cost

```python
# import numpy package for arrays and stuff 
import numpy as np  
  
# import matplotlib.pyplot for plotting our result 
import matplotlib.pyplot as plt   

dataset = np.array( 
[['Asset Flip', 100, 1000], 
['Text Based', 500, 3000], 
['Visual Novel', 1500, 5000], 
['2D Pixel Art', 3500, 8000], 
['2D Vector Art', 5000, 6500], 
['Strategy', 6000, 7000], 
['First Person Shooter', 8000, 15000], 
['Simulator', 9500, 20000], 
['Racing', 12000, 21000], 
['RPG', 14000, 25000], 
['Sandbox', 15500, 27000], 
['Open-World', 16500, 30000], 
['MMOFPS', 25000, 52000], 
['MMORPG', 30000, 80000] 
]) 


# select all rows by : and column 1 
# by 1:2 representing features 
X = dataset[:,1:2].astype(int) 
# targets
Y = dataset[:,2].astype(int)

# Plotting cost vs profit
x_plot=dataset[:,1].astype(int)
plt.plot(x_plot,Y)
plt.xlabel("Cost")
plt.ylabel("Profit")
plt.grid()
plt.show()
```

In the plot above we see the relation between cost and profit. Now our objective is to train our regressor and try to predict profit for a given cost

```python
from sklearn import linear_model

# initialize model
regr = linear_model.LinearRegression()
# train the model
regr.fit(X, Y)
# make predictions using the same features
predictions = regr.predict(X)
# plot out the predicted profits for the same features
plt.plot(X,predictions)
plt.xlabel("Cost")
plt.ylabel("Profit")
plt.grid()
plt.show()
```

Thus you see our linear regressor finds the closest linear fit to the data. It's only an approximation
